# Numpy Arrays

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

## Creating NumPy Arrays

### From a Python List

We can create an array by directly converting a list or list of lists:

```python
import numpy as np
l = [10,20,30,40,50] # 1D array
array = np.array(l)
print(array)

```

Lets create a matrix using list

```python
a = [[1,2,3],
     [4,5,6],
     [7,8,9]]
matrix = np.array(a)
matrix
```

## Built-in Methods

there are lots of build in methods to create an array lets create some and see

np.zeros(), np.ones(), np.full()

np.arange(), np.linspace()

np.eye()

```python
# np.zeros for 1D
np.zeros(4)
```

```python
# np.zeros for matrix
# Rows and column
np.zeros([3,4])




```

```python
# np.ones for 1D
np.ones(4)
```

```python
# np.ones for matrix
np.ones([3,4])
```

```python
# np.full for creating own...
np.full(5,7)
```

```python
# np.full for matrix
np.full([3,3],7)
```

```python
# np.arrange() for 1D
arr = np.arange(0,10,2) # using(start,stop,steps)
arr
```

```python
# np.linspace() start =0 , stop = 1, value = 5
np.linspace(0,1,5)

```

```python
# np.eye() Identity Matrix
np.eye(4)
```

```python
# np.random.rand() # rows,columns
np.random.rand(2,3)
```

```python
# np.random.randint(start,stop,(rows,columns))
np.random.randint(1,20,(3,3))
#""" randint :- 1 se 20 ke bech ki random value se matrix create karega """
```

arange → step-based values

linspace → fixed number of values

eye → identity matrix

rand → random floats (0–1)

randint → random integers in a range

randn → random values from bell curve distribution
