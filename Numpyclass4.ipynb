##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
# Aam zindegi
import numpy as np
array = np.array([2,10,56,1,54])
print(np.sort(array))


index = np.argsort(array)
print(index)

```

```python
# Matrix sort
matrix = np.array([[4,3,12],[14,10,15]])
#1.Row wise sort(axis = 1)
np.sort(matrix, axis=1)

#2. column Wise Sort(axis=0)
np.sort(matrix,axis=0)

#3. Sort Complete Matrix
np.sort(matrix)
```

##Concatenation & Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

np.concatenate([a,b]) # use for add
```

Vertical & Horizontal Stacking

```python
A = np.array([[1,2],[4,5]])
B = np.array([[7,8],[10,11]])

#Vertical Stacking
np.vstack([A,B])

#Horizontal Stacking
np.hstack([A,B])


```
