# Q1. Array Addition

## Problem Statement
Create two NumPy arrays:

```python
arr1 = [1, 2, 3, 4]
arr2 = [10, 20, 30, 40]
```

Add both arrays element-wise.

## Sample Output
```python
[11 22 33 44]
```


````python
import numpy as np
a = np.array([1,2,3,4])
b = np.array([10,20,30,40])
print(a+b)
````


# Q9. Row-wise Sum using Axis

## Problem Statement
Create the array:

```python
[[1,2,3],
 [4,5,6]]
```

Find row-wise sum using axis.

## Sample Output
```python
[ 6 15]
```


````python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
result = arr.sum(axis=1)
print(result)
````
